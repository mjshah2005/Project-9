```python
import pandas as pd
import matplotlib.pyplot as plt

def main():
    df = None

    while True:
        print("\nData Analysis Program")
        print("1. Load")
        print("2. Explore Data")
        print("3. Handle Missing Data")
        print("4. Generate Descriptive Statistics")
        print("5. Data Visualization")
        print("6. Save Visualization")
        print("7. Exit")
        
        choice = input("Select (1-7): ")

        if choice == '1':
            file = input("Enter file name: ")

            df = pd.read_csv(file)
            print("Loaded")

        elif choice == '2':
            print(df.head())

        elif choice == '3':
            # Shows rows that have any missing values
            print(df[df.isnull().any(axis=1)])

        elif choice == '4':
            print(df.describe())

        elif choice == '5':
            t = input("1. Bar 2. Line 3. Scatter: ")
            x = input("x Column: ")
            y = input("y Column: ")
            
            plt.figure() # Creates a new clean figure
            if t == '1':
                plt.bar(df[x], df[y])
            elif t == '2':
                plt.plot(df[x], df[y])
            elif t == '3':
                plt.scatter(df[x], df[y])
            
            plt.xlabel(x)
            plt.ylabel(y)
            
            plt.show()

        elif choice == '6':
            name = input("Save as (e.g., plot.png): ")
            plt.savefig(name)
            print("Saved")

        elif choice == '7':
            print("Exiting...")
            break

if __name__ == "__main__":
    main()
```
